***I will being visualizing the data from a fmri seaborn dataset.Fmri stands for Functional magnetic resonance imaging. It measures the small changes in blood flow with brain activity.***

```python
import plotly.express as px
import pandas as pd
import seaborn as sns
```

```python
fmri = sns.load_dataset("fmri")
fmri.head(20)
```

```python
#Create a line plot for event, region, and subjects.
px.line(fmri, x = 'timepoint', y = 'signal')
#The reason why graph looks like this is beacuse the values in the dataset is not sorted.
```

```python
#Sort the values using the sort_values() function in pandas
fmri.sort_values(by= ['subject','event','region','timepoint'], inplace=True)
```

```python
#the data has been sorted correctly.
fmri.head(20)
```

```python
#Create a line plot for event, region, and subjects.
px.line(fmri, x = 'timepoint', y = 'signal', facet_row='region',facet_col= 'event')
```

Shows the brain reaction to different events on the left and stimulations on the right over a timepoint.

```python
#The graph shows the different subjects and their brain activity.
px.line(fmri, x = 'timepoint', y = 'signal', facet_row='region',facet_col= 'event',color='subject',hover_name='subject')
```
